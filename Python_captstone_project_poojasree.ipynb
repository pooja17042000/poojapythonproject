```python
data.isnull().sum()
```

# Task 10

```python
top500_movies_revenue=data.sort_values(by='revenue',ascending=False).head(500)
top500_movies_revenue=top500_movies_revenue.loc[:,['revenue','budget','runtime','title']]
top500_movies_revenue
```

```python
import seaborn as sns
import statistics as st
```

```python
Mean=[]
Median=[]
Mode=[]
for col in top500_movies_revenue.head():
    Mean.append(top500_movies_revenue.loc[:,col].mean()) #calculating each column mean and stored within Mean named list
    Median.append(top500_movies_revenue.loc[:,col].median()) #calculating each column median and stored within Median named list
    Mode.append(st.mode(top500_movies_revenue[col])) #calculating each column mode and stored within Mode named list
    
row_head=['Mean','Median','Mode']
col_names=['revenue','budget','runtime']

#Creating a new dataframe with all calculated mean, median, and mode of 3 companies
d_data=pd.DataFrame([Mean, Median, Mode],columns=col_names)

#Inserting a new column 'Measures' at the begining of the dataframe with 3 measures names
d_data.insert(0, "Measures", row_head, True)

print(d_data)
```
